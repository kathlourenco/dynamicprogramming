```python
fila = []

for i in range(3):
    nome = input(f"digite o nome do cliente {i + 1}: ")
    fila.append(nome)
while fila:
    cliente = fila.pop(0)
    print(f"Atendendo: {cliente}")
```

```python
fila = []
i = 0
while True:
    doc = input(f"digite o documento {i + 1} (ou 'fim' para encerrar): ")
    if doc.lower() == "fim":
        break
    fila.append(doc)
    i += 1
while fila:
    ordem = fila.pop(0)
    print(f"imprimindo: {ordem}")
```

```python
fila = ["kath", "clara", "luis", "leo", "sara"]

while fila:
    atend = fila.pop(0)
    print(f"atendendo: {atend}")
    print(f"pessoas na fila: {len(fila)}")
```

```python
fila = []
senha = 1

while True:
    nome = input("nome do cliente (ou 'fim' para parar): ")
    if nome.lower() == "fim":
        break
    fila.append((senha, nome))
    print(f"senha{senha} gerada para {nome}")
    senha += 1
while fila:
    numero, cliente = fila.pop(0)
    print(f"chamando a senha {numero} - {cliente}")
```

```python
fila = []

while True:
    nome = input("nome do cliente (ou 'fim' para parar): ")
    if nome.lower() == "fim":
        break
    tempo = int(input("tempo de preparo (min): "))
    fila.append((nome, tempo))

tempoTotal = 0

while fila:
    cliente, tempo = fila.pop(0)
    print(f"atendendo: {cliente} — preparo: {tempo} min")
    tempoTotal += tempo

print(f"\ntempo total de preparo: {tempoTotal} min")
```

```python
fila = []

while True:
    print("\n1 - adicionar cliente")
    print("2 - atender cliente")
    print("3 - mostrar fila")
    print("4 - sair")
    opcao = input("escolha: ")

    if opcao == "1":
        nome = input("nome do cliente: ")
        fila.append(nome)
        print(f"'{nome}' adicionado")

    elif opcao == "2":
        if fila:
            atendido = fila.pop(0)
            print(f"atendendo: {atendido}")
        else:
            print("fila vazia")

    elif opcao == "3":
        if fila:
            print(f"fila atual: {fila}")
        else:
            print("fila vazia")

    elif opcao == "4":
        print("encerrando")
        break

    else:
        print("opção invalida")
```
