```python
pilha = []

for i in range(3):
    nome = input(f"digite o nome {i + 1}: ")
    pilha.append(nome)

while pilha:
    retirado = pilha.pop()
    print(f"retirando: {retirado}")
```

```python
pilha = []

while True:
    doc = input("digite o documento (ou 'fim' para parar): ")
    if doc.lower() == "fim":
        break
    pilha.append(doc)

while pilha:
    processado = pilha.pop()
    print(f"processando: {processado}")

print("todos os docs foram processados")
```

```python
historico = []

while True:
    url = input("digite a url (ou 'voltar' / 'fim'): ")

    if url.lower() == "fim":
        if historico:
            print(f"pagina atual: {historico[-1]}")
        else:
            print("nenhuma pagina visitada")
        break

    elif url.lower() == "voltar":
        if historico:
            saiu = historico.pop()
            print(f"saindo de: {saiu}")
            if historico:
                print(f"pagina atual: {historico[-1]}")
            else:
                print("nenhuma pagina anterior")
        else:
            print("historico vazio")

    else:
        historico.append(url)
        print(f"visitando: {url}")
```

```python
acoes = []

while True:
    entrada = input("digite uma ação (ou 'desfazer' / 'fim'): ")

    if entrada.lower() == "fim":
        break

    elif entrada.lower() == "desfazer":
        if acoes:
            desfeita = acoes.pop()
            print(f"acao desfeita: {desfeita}")
        else:
            print("nenhuma acao para desfazer")

    else:
        acoes.append(entrada)
        print(f"acao registrada: {entrada}")

print(f"\nacoes restantes: {acoes}")
```
